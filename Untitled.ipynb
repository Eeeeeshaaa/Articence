```python
DATA_DIR = "data"

documents = []
for file in os.listdir(DATA_DIR):
    if file.endswith(".pdf"):
        loader = PyPDFLoader(os.path.join(DATA_DIR, file))
        documents.extend(loader.load())

print("Total pages loaded:", len(documents))
```

```python
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=80
)

chunks = text_splitter.split_documents(documents)
print("Total chunks:", len(chunks))
```

```python
tts_engine = pyttsx3.init()
tts_engine.setProperty("rate", 175)

def speak(text):
    tts_engine.say(text)
    tts_engine.runAndWait()
```

```python
import os
import whisper
import sounddevice as sd
from scipy.io.wavfile import write

asr_model = whisper.load_model("base", device="cpu")

def listen(seconds=5):
    fs = 16000

    print("Recording... Speak now.")
    recording = sd.rec(
        int(seconds * fs),
        samplerate=fs,
        channels=1,
        dtype="float32"
    )
    sd.wait()
    print("Recording finished.")

    audio_path = os.path.abspath("input.wav")

    write(audio_path, fs, recording)

    print("Saved at:", audio_path)
    print("File exists:", os.path.exists(audio_path))

    audio = whisper.load_audio(audio_path)
    audio = whisper.pad_or_trim(audio)

    result = asr_model.transcribe(audio, fp16=False)

    print("Transcribed Text:", result["text"])

    return result["text"]
```

```python
embedding_model = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

vectorstore = FAISS.from_documents(chunks, embedding_model)
vectorstore.save_local("vector_store")
```

```python
corpus = [c.page_content for c in chunks]
tokenized_corpus = [doc.lower().split() for doc in corpus]

bm25 = BM25Okapi(tokenized_corpus)
```

```python
reranker = CrossEncoder("cross-encoder/ms-marco-MiniLM-L-6-v2")
```

```python
conversation_history = []

def rewrite_query(query):
    if "second" in query.lower() and conversation_history:
        return conversation_history[-1] + " " + query
    return query
```

```python
def hybrid_search(query, top_k=5):
    rewritten_query = rewrite_query(query)

    dense_docs = vectorstore.similarity_search(rewritten_query, k=10)
    dense_texts = [d.page_content for d in dense_docs]

    bm25_scores = bm25.get_scores(rewritten_query.lower().split())
    bm25_top_indices = sorted(
        range(len(bm25_scores)),
        key=lambda i: bm25_scores[i],
        reverse=True
    )[:10]

    bm25_texts = [corpus[i] for i in bm25_top_indices]

    candidates = list(set(dense_texts + bm25_texts))

    pairs = [(rewritten_query, c) for c in candidates]
    scores = reranker.predict(pairs)

    ranked = sorted(zip(candidates, scores), key=lambda x: x[1], reverse=True)
    return ranked[:top_k]
```

```python
def to_spoken_english(text):
    text = text.replace("CPU", "C P U")
    text = text.replace("RAM", "R A M")
    text = text.replace("GHz", "gigahertz")
    text = text.replace("e.g.", "for example")

    sentences = text.split(".")
    return ". ".join(sentences[:3])
```
